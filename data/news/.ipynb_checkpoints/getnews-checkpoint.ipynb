```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.sozcu.com.tr/ekonomi"
url_2 = "https://www.sozcu.com.tr/gundem"

headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
response_2 = requests.get(url_2, headers=headers)

soup = BeautifulSoup(response.text, "html.parser")
soup_2 = BeautifulSoup(response_2.text, "html.parser")

sozcu_news_economi = soup.find_all("span", class_="small text-secondary text-truncate-2")
sozcu_news_current  = soup_2.find_all("span", class_="small text-secondary text-truncate-2")

sozcu_news = sozcu_news_economi + sozcu_news_current

for b in sozcu_news:
    print(b.text.strip())
```

```python
url = "https://www.ntv.com.tr/ekonomi"
url_2 = "https://www.ntv.com.tr/turkiye"

headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
response_2 = requests.get(url_2, headers=headers)

soup = BeautifulSoup(response.text, "html.parser")
soup_2 = BeautifulSoup(response_2.text, "html.parser")

ntv_economi = soup.find_all("a", class_="card-text-link text-elipsis-3")
ntv_news_current  = soup_2.find_all("a", class_="card-text-link text-elipsis-3")

ntv_news = ntv_economi + ntv_news_current

for b in ntv_news:
    baslik = b.text.strip()
    print(baslik)
```
